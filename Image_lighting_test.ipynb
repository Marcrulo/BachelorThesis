```python
import os
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt

from skimage import io, color, exposure
from skimage.filters import prewitt, median, gaussian
from skimage.morphology import disk, dilation, erosion, closing, opening 
from skimage.transform import rescale

from sklearn.decomposition import PCA, IncrementalPCA

from pystackreg import StackReg
```

```python
img_path = os.path.abspath(os.path.join(
    os.getcwd(),
    os.pardir,
    'camera&image',
    'lighting_test2', # lighting_test2
))
```

```python
imgs = list(io.imread_collection(img_path+'/*.bmp'))
```

```python
for img in imgs:
    plt.imshow(img, cmap='gray')
    plt.show()
```

```python
figs, axs = plt.subplots(2,4,figsize=(20,10))
i = 0
for row in range(2):
    for col in range(4):
        axs[row, col].hist(imgs[i].ravel(), bins=20)
        axs[row, col].set_ylim(0,220000)
        axs[row, col].grid()
        i += 1

axs[0,0].set_title("ceiling light, seated")
axs[0,1].set_title("ceiling light, standing")
axs[0,2].set_title("ceiling light, seated, pc blocked")
axs[0,3].set_title("ceiling light, standing, pc blocked")

axs[1,0].set_title("no light, open door")
axs[1,1].set_title("no light, closed door")
axs[1,2].set_title("no light, open door, pc blocked")
axs[1,3].set_title("no light, closed door, pc blocked")

plt.show()
```
