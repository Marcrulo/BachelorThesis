```python
import pandas as pd
import os
from matplotlib import pyplot as plt
```

```python
csv_path = os.path.abspath(os.path.join(
    os.getcwd(),
    os.pardir,
    'datasets',
    'quality_scores.csv'
))


df = pd.read_csv(csv_path)
df
```

```python
filter_df = df.copy()
#filter_df = filter_df[filter_df['Param1']>1.2]
#filter_df = filter_df[filter_df['Area sum']>20000]

fig, axs = plt.subplots(1,4,figsize=(20,5))
for i in range(4):
    axs[i].scatter(filter_df["Param"+str(i+1)], filter_df["Area sum"])
    # axs[i].set_title("Param"+str(i+1))
    axs[i].set_ylim(df['Area sum'].min(), df['Area sum'].max()+500)
axs[0].set_title("Proportion Valve Porex")
axs[1].set_title("Weld Energy")
axs[2].set_title("Weld Pressure")
axs[3].set_title("Triggerforce")

plt.show()
```
